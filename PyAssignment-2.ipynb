```python
sns.scatterplot(x="total",y="speeding",data=df)
```

* Inference : Total drivers and number of car crashes due to speeding are increasing , it insinuates proportionally but not totally proportional

```python
sns.scatterplot(x="total",y="no_previous",data=df)
```

* Inference : **Directly proportional** - As total drivers are increasing , car crashes due to no previous experience are also increasing

```python
sns.lineplot(x="ins_losses",y="no_previous",data=df)
```

* Inference : It was both increasing and decreasing at times. The lowest point has occured at 137  (ins_losses) and highest at 143 (ins_losses) [Approx]

```python
sns.scatterplot(x="alcohol",y="speeding",data=df)
```

:* Inference : Directly proportional ( Alcohol and Speeding)

```python
sns.distplot(df["ins_premium"])
```

* Inference : Cars with insurence premium around 800 were crashing more whereas cars with insurance premium around 1100 ~ 1300 are less likely to crash.

```python
sns.relplot(x="total",y="speeding",data=df,hue="abbrev")
```

```python
plt.figure(figsize=(17, 6))
sns.barplot(x="abbrev",y="speeding",data=df)
plt.show()
```

* Inference : In the HI (Hawaii State in US) there are more % of overspeeding drivers who are responsible for car crashes

```python
boxplot_for = df[['total', 'speeding']]
plt.figure(figsize=(7, 8))
sns.boxplot(data=boxplot_for)
plt.show()
```

* Inference : From the above boxplot , we can see a outliner between 9 and 11 (approximately)

```python
plt.figure(figsize=(12, 8))
sns.histplot(x="total",data=df)
plt.show()
```

* Inference: At 12.5 the count reached its peak

```python
plt.figure(figsize=(17, 12))
sns.jointplot(x="ins_premium",y="ins_losses",data=df)
plt.show()
```

* Inference: As the ins_premiums increases the ins_losses are also increasing (Nearly Directly proportional). This is a graph of combination of bivariate and univariate

Correlation:

* ">0.5" - Highly correlated
* "<0.5" - less correlated
* "=0.5" - neutral

```python
correlation_value = df.corr(numeric_only=True)
correlation_value
```

* Inference : From the corr() we can find all corellations values for each with other parameter how it was related

```python
df[['total','speeding']].corr()
```

```python
sns.heatmap(correlation_value,annot=True)
```

Inference :
* Less correlated : Total & ins_premium
* Neutrally correlated : None
* Highly correlated : Total & no_previous
